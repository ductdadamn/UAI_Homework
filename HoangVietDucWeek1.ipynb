Câu 1 – 1D Array

Tạo một numpy array 1 chiều chứa các số: 1, 2, 3.

In mảng ra màn hình

In số chiều của mảng

Tạo một numpy array 2 chiều từ danh sách:

[(1,2,3),
 (4,5,6)]


In mảng

In ndim, shape, size

```python
b=np.array([[1,2,3], [4,5,6]])
print(b)
print(b.ndim)
print(b.shape)
print(b.size)
```

Câu 4 – So sánh bộ nhớ

Viết chương trình:

Tạo list từ 0 → 999

Tạo numpy array từ 0 → 999

In lượng bộ nhớ dùng cho list và numpy array

Comment: cái nào tiết kiệm hơn?

```python
import sys
a=list(range(1000))
b=np.array(range(0,1000))
print(sys.getsizeof(a))
print(b.nbytes)
#Numpy array is more economical than list
```

Câu 5 – Reshape

a = np.array([(8,9,10),(11,12,13)])

Câu 6 – Slicing cơ bản

Cho mảng:

a = np.array([(1,2,3,4),
              (3,4,5,6)])


Lấy phần tử 3

Lấy toàn bộ phần tử ở cột thứ 3

Giải thích bằng comment từng dòng slicing

```python
a = np.array([(1,2,3,4), (3,4,5,6)])
print(a[0,2])#3 is at 1st row and 3rd column so its index is 0 and 2
print(a[1,0])#3 is at 2nd row and 1st column so its index is 1 and 0
print(a[:,2])#3rd column means its index is 2, : to remove other positions

```

Câu 8 – min, max, sum

a = np.array([1,2,3])

Lớn nhất

Tổng các phần tử

Câu 9 – axis

Cho mảng:

a = np.array([(1,2,3),
              (3,4,5)])


Tính tổng theo cột

Tính tổng theo hàng

Giải thích axis=0 và axis=1 bằng comment

```python
a = np.array([(1,2,3), (3,4,5)])
print(a.sum(axis=0))#column
print(a.sum(axis=1))#row
```

Câu 10 – Các phép toán cơ bản

Cho:

x = np.array([(1,2,3),(3,4,5)])
y = np.array([(1,2,3),(3,4,5)])


Cộng

Trừ

Nhân

Chia

Dot product

```python
x = np.array([(1,2,3),(3,4,5)])
y = np.array([(1,2,3),(3,4,5)])
print(x+y)
print(x-y)
print(x*y)
print(x/y)
y=y.reshape(3,2)
print(np.dot(x,y))
```
